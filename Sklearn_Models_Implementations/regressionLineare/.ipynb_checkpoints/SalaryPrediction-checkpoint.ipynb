```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("E:\\MachineLearningCoursera\\MachineLeaningPractice\\datasets\\salary_data.csv")
```

```python
data.head()
```

```python
data.info()
```

# Preparing The Data

```python
X=data[['YearsExperience']]
Y=data['Salary']
```

```python
plt.scatter(X,Y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(X,Y,test_size=0.33,random_state=0)
```

# Our Model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
print(f" le poid : {regressor._coef}")
print(f" l'odonne a l'origine : {regressor.intercept}")
```

```python
ordonne = np.linspace(0,15,1000)
plt.scatter(X,Y)
plt.plot(ordonne ,ordonne* regressor.coef_[0]+regressor.intercept_,c="r")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

# Evaluate

```python
y_predict = regressor.predict(x_test)
```

```python
y_predict
```

```python
from sklearn import metrics
```

```python
print(f" MAE : {metrics.mean_absolute_error(y_predict,y_test)} ")
print(f"MSE : {metrics.mean_squared_error(y_predict,y_test)}")
print(f"RMSE : {np.sqrt(metrics.mean_squared_error(y_predict,y_test))}")
print(f" R2 : {metrics.r2_score(y_predict,y_test)}")
```
